```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

data_import = pd.read_csv("LasVegasTripAdvisorReviews-Dataset.csv")
LasVegas_Reviews = pd.DataFrame(data_import)

LasVegas_Reviews_Grouping = LasVegas_Reviews.groupby(["User country"])["Nr. reviews"].sum()
LasVegas_Reviews_Grouping.plot(kind = "bar", x = "User country", y = "Nr. reviews")
plt.title("Number of Reviews by Country")
plt.xlabel("Country")
plt.ylabel("Numbe of Reviews")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

data_import = pd.read_csv("LasVegasTripAdvisorReviews-Dataset.csv")
LasVegas_Reviews = pd.DataFrame(data_import)

LasVegas_Reviews_Grouping_NrReviews = LasVegas_Reviews.groupby(["User country"])["Nr. reviews"].sum()
LasVegas_Reviews_Grouping_NrHotelReviews = LasVegas_Reviews.groupby(["User country"])["Nr. hotel reviews"].sum()
LasVegas_Reviews_Grouping_AvgHotelReviews = LasVegas_Reviews_Grouping_NrHotelReviews / LasVegas_Reviews_Grouping_NrReviews
LasVegas_Reviews_Grouping_AvgHotelReviews.plot(kind = "bar", color = "green")
plt.xlabel("Country")
plt.ylabel("%age of Reviews that are Hotel Reviews")
plt.title("Hotel Reviews / Total Reviews by Each Country")
```
